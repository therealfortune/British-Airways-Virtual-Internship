```python
from bs4 import BeautifulSoup
import requests 
import re
import csv 
import time
```

```python
url = 'https://www.airlinequality.com/airline-reviews/british-airways/?sortby=post_date%3ADesc&pagesize=20'
f = requests.get(url).text 

soup = BeautifulSoup(f, 'html.parser')
```
